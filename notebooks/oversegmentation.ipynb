```python
from scipy import ndimage as ndi
import mahotas as mh
import numpy as np
import glob
import os
import re
import cv2
from skimage.segmentation import felzenszwalb, random_walker, slic, quickshift, mark_boundaries
from skimage import img_as_float
import matplotlib.pyplot as plt 
%matplotlib inline

WS_PARAMS = [ 'SIGMA_WS', 'MASK_SIZE' ]
FZ_PARAMS = [ 'SCALE', 'SIGMA', 'MIN_SIZE' ]
CWS_PARAMS = [ ]
RW_PARAMS = [ 'BETA', 'TOLERANCE']


def plot_pair(img1, title1, img2, title2):
    plt.subplot(121), plt.imshow(img1)
    plt.title(title1), plt.xticks([]), plt.yticks([])
    plt.subplot(122), plt.imshow(img2)
    plt.title(title2), plt.xticks([]), plt.yticks([])
    plt.show()

def plot_figure(img, title=None, colorbar=False):
    plt.figure()
    plt.imshow(img)
    if colorbar is True:
        plt.colorbar()
    if title is not None:
        plt.title(title)

""" Checks all parameters exist in the dictionary """


def check_params(param_keys, params):
    for key in param_keys:
        if not key in params:
            return False
    return True


""" Thresholds image. The type 't' of thresholding can be: 
    - ADAPTATIVE. It uses the local information. Params:
        - param1: local area to consider
        - param2: weight for neighboring information
        - param3: if true, adaptative threhsolding performed using gaussian mask.
    - GLOBAL. Computes a global threshold using Otsu's algorithm. Params [1-3] are ignored.
    - CUT. Provides  (histogram cut with Otsu algorithm).
        - param1: minimum value for preserving foreground. """


def threshold(img, t, inv, param1, param2, param3):
    thresh_binary = cv2.THRESH_BINARY_INV if inv is True else cv2.THRESH_BINARY
    if t == 'ADAPTATIVE':
        local_type = cv2.ADAPTIVE_THRESH_GAUSSIAN_C if param3 is True \
            else cv2.ADAPTIVE_THRESH_MEAN_C 
        thresh = cv2.adaptiveThreshold(img, 255, local_type, thresh_binary, param1, param2)
    elif t == 'GLOBAL':
        ret1, thresh = cv2.threshold(img, 0, 255, thresh_binary+cv2.THRESH_OTSU)
    elif t == 'CUT':
        ret1, thresh = cv2.threshold(img, param1, 255, thresh_binary)
    else:
        raise Exception("Unknown type {}:" + t)
    return thresh


""" Computes the watershed algorithm after eliminating sure foreground and background content
Instructions: http://docs.opencv.org/3.1.0/d3/db4/tutorial_py_watershed.html#gsc.tab=0"""


def complex_watershed_segmentation(f):
   
    # Threshold image - Main step
    # Values below threshold are discarded and mapped into thresh_max
    # OTSU algorithm automatically ocmputes the global threhsold by maximizing the inter-class variance
    # between background and foreground pixels. If OTSU used, 'thresh_min' is not used. foreground pixels
    # are set to the value of 'thresh_max'. Important to set a high value of thresh_max (still do not 
    # know why but it achieves less merges)
    # ret1, thresh = cv2.threshold(f, thresh_min, thresh_max, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
    thresh = threshold(f, 'GLOBAL', True, 3, 0, True)
    plot_figure(thresh, 'threhsold', True)

    # Sure background area - Mask must be small ~ order of 3
    kernel = np.ones((3, 3), np.uint8)
    sure_bg = cv2.dilate(thresh, kernel, iterations=1)
    plot_figure(sure_bg, 'sure background', True)

    # Sure foreground area 
    # Given the thresholded image, it computes the distance of each point to the closest
    # background region. The closer the pixels is to the border, the weaker its intensity
    dist_transform = cv2.distanceTransform(thresh, cv2.cv.CV_DIST_L1, 3)
    plot_figure(dist_transform, 'distance transform')
    
    # Threshold distance transform to delete weaker points
    # Add parameter for threhsold of distance transform - maybe we could add a model for this part!
    dist_transform = np.uint8(dist_transform)
    sure_fg = threshold(dist_transform, 'CUT', False, 0.015 * dist_transform.max(), 0, True)
    plot_figure(sure_fg, 'sure foreground', True)

    # Finding unknown region
    sure_fg = np.uint8(sure_fg)
    plot_figure(sure_fg, 'sure foreground (uint8)', True)
    unknown = cv2.subtract(sure_bg, sure_fg)
    plot_figure(unknown, 'Unknown regions', True)

    # Marker labelling. We use connected components function that labels the image regions
    # using a connectivity mask. It seems it is better to use a simpler mask to avoid 
    # over-connection of regions
    (markers, ret3) = ndi.label(sure_fg)
    
    # Add one to all labels so that sure background is not 0, but 1
    markers += 1
    # Now, mark the region of unknown with zero
    markers[unknown == 255] = 0
    markers.astype(float)

    # Apply watershed
    watershed = mh.cwatershed(f, markers)

    # Assign background again. Check
    watershed -= 1

    return watershed


""" Segmentation using random walker technique """

def random_walker_segmentation(f, beta, tol):
    segments = random_walker(f, markers, beta=50, mode='cg', tol=0.1)
    return segments 


""" Superpixel formation using the Felzenszwalb segmentation algorithm """


def felzenszwalb_segmentation(f, scale, sigma, min_size):
    segments_fz = felzenszwalb(f, scale=scale, sigma=sigma, min_size=min_size)
    return segments_fz


""" Superpixel formation by using watershed on the local minimas 
after a Gaussian blurring"""


def watershed_segmentation(f, sigma_ws, mask_size):
        
    # Threshold image
    value, thresh_img = cv2.threshold(f, 0, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
    
    # Compute distance transform maximas
    dist_transform = cv2.distanceTransform(thresh_img, cv2.cv.CV_DIST_L2, 3)
    seeds = mh.regmax(dist_transform, np.ones((mask_size,mask_size)))
    
    # Connect component
    (markers, num_seeds) = ndi.label(seeds)
    
    # Gaussian blur
    membrane_watersheds = mh.gaussian_filter(f, sigma_ws)
    
    # Uses watershed algorithm for segmentation (fills basins)
    # Regions with adjacent catchment basins are constructed.
    # Usually produces oversegmentation of the image. Works on the
    # gradient image
    ws = mh.cwatershed(membrane_watersheds, markers)
    return ws


""" Computes the superpixels generation performance on the given image
for the input method and parameters"""


def compute_superpixels(f, method, params):

    membrane = mh.imread(f)

    if method == 'watershed':
        if check_params(WS_PARAMS, params):
            return watershed_segmentation(membrane, params[WS_PARAMS[0]],
                params[WS_PARAMS[1]])
        else :
            raise Exception("Wrong parameters for the {} method".format(method))
    elif method == 'complex_watershed':
            return complex_watershed_segmentation(membrane)
    elif method == 'felzenszwalb':
        if check_params(FZ_PARAMS, params):
            return felzenszwalb_segmentation(membrane, params[FZ_PARAMS[0]], params[FZ_PARAMS[1]],
                params[FZ_PARAMS[2]])
        else :
            raise Exception("Wrong parameters for the {} method".format(method))
    elif method == 'random_walker':
        if check_params(RW_PARAMS, params):
            return random_walker_segmentation(membrane, params[RW_PARAMS[0]], params[RW_PARAMS[1]])
        else :
            raise Exception("Wrong parameters for the {} method".format(method))
    else:
        raise Exception("Unkown method '{}'".format(method))


if __name__ == "__main__":

    # method = 'watershed'
    # params = { 'SIGMA_WS': 5, 'MASK_SIZE': 5 }
    
    # method = 'random_walker'
    # params = { 'BETA': 25, 'TOLERANCE': 5}

    #method = 'felzenszwalb'
    #params = { 'SCALE': 185, 'SIGMA': 0.3, 'MIN_SIZE': 40}
    
    method = 'complex_watershed'
    params = {}
    
    result = compute_superpixels('/home/dmora/neural/snemi/00_prepare_data/membrane/00000.png', method, params)
    
    # Store and show prediction
    #mh.imsave('/tmp/guest-P6APGC/Desktop/watershed_jan.png', result.astype(numpy.uint8))
    plot_figure(result, 'Prediction', True)
    mh.imsave('/home/dmora/Desktop/a.tif', result.astype(float))
    
    # Show groundtruth 
    #truth = mh.imread('/tmp/guest-P6APGC/neuron/snemi/00_prepare_data/groundtruth/neuron_ids0000.png')
    #plot_figure(truth, 'Groundtruth', True)
```
