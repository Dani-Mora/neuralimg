```python
import cv2
import numpy as np
import pandas
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
matplotlib.style.use('ggplot')
%matplotlib inline

def plot_column(data, column):
    pivoted = data.pivot(index='MERGE_THRESH', columns='SETTING', values=column)
    sns.set_palette("hls", len(list(pivoted.columns.values)))
    ax = pivoted.plot(figsize=(10,8));
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., prop={'size':16})
    ax.set_ylabel(column)
```

```python
# Load data using Pandas
data = pandas.read_csv('z',
                          sep=" +", engine='python')
# frames = [data_sm, data_ext, data_extra]
# data = pandas.concat(frames)
# Use label for setting of sigma + mask_size
str_sigma = data["SIGMA_WS"].map(str)
str_mask = data["MASK_SIZE"].map(str)
data['SETTING'] = 'sigma(' + str_sigma + ')_mask(' + str_mask + ')'
data1 = data[data['SIGMA_WS'] <= 1.0]
data2 = data[data['SIGMA_WS'] >= 3.0]
```

```python
plot_column(data, 'TED_FM')
```

```python
plot_column(data, 'TED_FS')
```

```python
# data['TED_TOTAL'] = data['TED_FM'] + data['TED_FS']
# plot_column(data, 'TED_TOTAL')
data1['TED_TOTAL'] = data1['TED_FM'] + data1['TED_FS']
plot_column(data1, 'TED_TOTAL')
data2['TED_TOTAL'] = data2['TED_FM'] + data2['TED_FS']
plot_column(data2, 'TED_TOTAL')
```

```python
# data['TED SPLITS - TED FM * 10'] = data['TED_FS'] - data['TED_FM'] * 10
# plot_column(data, 'TED SPLITS - TED FM * 10')
data1['TED WEIGHTED'] = data1['TED_FS'] + data1['TED_FM'] * 10
plot_column(data1, 'TED WEIGHTED')
data2['TED WEIGHTED'] = data2['TED_FS'] + data2['TED_FM'] * 10
plot_column(data2, 'TED WEIGHTED')
```

```python
plot_column(data, 'RAND')
```

```python
# plot_column(data, 'VOI_MERGE')
plot_column(data1, 'VOI_MERGE')
plot_column(data2, 'VOI_MERGE')
```

```python
# plot_column(data, 'VOI_SPLIT')
plot_column(data1, 'VOI_SPLIT')
plot_column(data2, 'VOI_SPLIT')
```

```python
# data['VOI WEIGHTED'] = data['VOI_SPLIT'] + data['VOI_MERGE'] * 10 
# plot_column(data, 'VOI WEIGHTED')
data1['VOI WEIGHTED'] = data1['VOI_SPLIT'] + data1['VOI_MERGE'] * 10 
plot_column(data1, 'VOI WEIGHTED')
data2['VOI WEIGHTED'] = data2['VOI_SPLIT'] + data2['VOI_MERGE'] * 10 
plot_column(data2, 'VOI WEIGHTED')
```

```python
# plot_column(data, 'NUM')
plot_column(data1, 'NUM')
plot_column(data2, 'NUM')
```
